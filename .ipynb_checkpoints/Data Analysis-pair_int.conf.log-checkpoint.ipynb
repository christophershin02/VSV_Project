```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating the dataframe for 1E9 antibody
df1 = pd.read_csv("1E9_Dataframe.txt", sep='\s+', engine='python')
df1['PoseNum_ClusterNum'] = df1['PoseNum_ClusterNum'].str.replace('/Users/chris/Documents/ASU_Work/Research_Singharoy/Minimization', '')
df1 = df1.drop(columns=['ETITLE:','BOUNDARY','MISC'])
df1
#Energy in kcal/mol
```

```python
### NOTES: Not much range of the potential energy
```

```python
#Sorting based on potential energy
df1 = df1.sort_values(by = "POTENTIAL")
```

```python

df1.head(n=50)
```

```python
df1['POTENTIAL'].hist()

plt.xlabel('Energy (kcal/mol)')
plt.ylabel('Frequency')
plt.title('Potential Energy')


plt.show()
```
